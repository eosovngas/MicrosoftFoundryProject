# Importar Librerias

````python
import os
from dotenv import load_dotenv
from openai import OpenAI
````

# Leer credenciales archivo .env

````python
load_dotenv()
endpoint = os.getenv("FOUNDRY_PROJECT_ENDPOINT")
endpoint2 = os.getenv("FOUNDRY_PROJECT_ENDPOINT2")
api_key = os.getenv("KEY_FOUNDRY_PROJECT_ENDPOINT")

print("FOUNDRY_PROJECT_ENDPOINT:", endpoint)
print("FOUNDRY_PROJECT_ENDPOINT:", endpoint2)
print("KEY_FOUNDRY_PROJECT_ENDPOINT:", api_key)
````

# Consumir modelo desde microsoft foundry empleando api key

````python
client = OpenAI(
    base_url=endpoint,
    api_key=api_key
)

response = client.responses.create(
    model="gpt-4.1",
    input="que sabes de chile",
)


# inspeccionar metadata
print(f"respuesta: {response.output[0]}")
# imprimir solo el texto de la respuesta
print(response.output_text)
````
